```python
income_dict = {'1':'Under $1', '2':'Between 1 and $10,000', '3':'Between 10,000 and $25,000', '4':'Between 25,000 and $50,000', '5':'Between 50,000 and $75,000', '6':'Between 75,000 and $100,000', '7':'Between 100,000 and $200,000', '8':'$200,000 or more'}
```

```python
TN_income_edit['income_bucket'] = TN_income_edit['income_bucket'].astype(str)
TN_income_edit['income_bucket'] = TN_income_edit['income_bucket'].replace(income_dict)
TN_income_edit.head()
```

```python
income_county_agg = TN_income_edit.groupby(by=['county']).sum().reset_index()
income_county_agg.head()
```

```python
income_county_agg_county_only = income_county_agg.loc[income_county_agg['county'] != 'Tennessee'] 
income_county_agg_county_only.loc[income_county_agg_county_only['return_count'].idxmax()]
```

```python
income_county_agg_county_only.loc[income_county_agg_county_only['agi'].idxmax()]
```

```python
TN_income_total = TN_income_edit.loc[TN_income_edit['county'] == 'Tennessee']
TN_income_total.head()
```

```python
import matplotlib.pyplot as plt
plt.barh('income_bucket', 'return_count', data=TN_income_total, color = 'red', edgecolor = 'black')
plt.xlabel('Returns')
plt.title('Total Returns by Income in TN');
```

```python
income_county_agg['mean_income'] = 1000*(income_county_agg['total_inc_amt'] / income_county_agg['return_count'])
income_county_agg.head()
```

```python
income_county_agg_county_only = income_county_agg.loc[income_county_agg['county'] != 'Tennessee']
income_county_agg_county_only.head()
```

```python
plt.hist('mean_income', data=income_county_agg_county_only, color='blue', edgecolor='black', bins=14, range = (0, 140000))
plt.xlabel('Mean Income')
plt.ylabel('Number of Counties')
plt.title('Distribution of Mean Income in TN Counties');
```

```python
ha_data_boxplot = [ha_costs_df_urban['analysis_value'], ha_costs_df_rural['analysis_value']]
labels=TN_counties_ha['urban'].unique()
plt.boxplot(ha_data_boxplot, labels=labels)
plt.ylabel('Heart Attack Costs')
plt.title('Distribution of Heart Attack Costs in TN Counties');
```
